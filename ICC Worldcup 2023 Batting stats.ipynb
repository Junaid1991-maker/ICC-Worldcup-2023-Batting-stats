```python
# First, let's check the data types
print(df.dtypes)

# Convert numeric columns that might be stored as strings
numeric_cols = ['Runs', 'Average', 'SR', '100', '50']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # coerce will turn invalid values to NaN

# Now let's verify the conversions
print("\nAfter conversion:")
print(df[numeric_cols].dtypes)

# Handle any missing values if needed (from coercion)
print("\nMissing values after conversion:")
print(df[numeric_cols].isnull().sum())

# Now run the team stats analysis again
team_stats = df.groupby('Country').agg({
    'Runs': 'sum',
    'Average': 'mean',
    'SR': 'mean',
    '100': 'sum',
    '50': 'sum'
}).sort_values('Runs', ascending=False)

# Plot team runs
plt.figure(figsize=(12, 6))
sns.barplot(x=team_stats.index, y=team_stats['Runs'], palette='viridis')
plt.title('Total Runs by Team')
plt.xticks(rotation=45)
plt.ylabel('Total Runs')
plt.show()

# Plot team averages and strike rates
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(x=team_stats.index, y=team_stats['Average'], palette='coolwarm', ax=axes[0])
axes[0].set_title('Team Batting Averages')
axes[0].set_xticklabels(team_stats.index, rotation=45)
axes[0].set_ylim(0, team_stats['Average'].max()*1.1)  # Add some headroom

sns.barplot(x=team_stats.index, y=team_stats['SR'], palette='coolwarm', ax=axes[1])
axes[1].set_title('Team Strike Rates')
axes[1].set_xticklabels(team_stats.index, rotation=45)
axes[1].set_ylim(0, team_stats['SR'].max()*1.1)  # Add some headroom

plt.tight_layout()
plt.show()
```

 Top Performers by Key Metrics

```python
# Create a function to display top performers
def top_performers(df, metric, n=5, ascending=False):
    return df.sort_values(metric, ascending=ascending)[['Player', 'Country', metric]].head(n)

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 15))

# Top by Runs
top_runs = top_performers(df, 'Runs')
sns.barplot(x='Runs', y='Player', data=top_runs, palette='Blues_d', ax=axes[0])
axes[0].set_title('Top 5 Players by Runs')
axes[0].set_xlabel('Total Runs')

# Top by Average (min 5 innings)
min_innings = 5
top_avg = df[df['Inning'] >= min_innings].sort_values('Average', ascending=False).head(5)
sns.barplot(x='Average', y='Player', data=top_avg, palette='Greens_d', ax=axes[1])
axes[1].set_title(f'Top 5 Players by Average (min {min_innings} innings)')
axes[1].set_xlabel('Batting Average')

# Top by Strike Rate (min 100 balls faced)
min_balls = 100
top_sr = df[df['Ball_Faced'] >= min_balls].sort_values('SR', ascending=False).head(5)
sns.barplot(x='SR', y='Player', data=top_sr, palette='Reds_d', ax=axes[2])
axes[2].set_title(f'Top 5 Players by Strike Rate (min {min_balls} balls faced)')
axes[2].set_xlabel('Strike Rate')

plt.tight_layout()
plt.show()

# Display tables for each category
print("Top 5 Players by Runs:")
display(top_runs)

print(f"\nTop 5 Players by Average (min {min_innings} innings):")
display(top_avg[['Player', 'Country', 'Average', 'Inning']])

print(f"\nTop 5 Players by Strike Rate (min {min_balls} balls faced):")
display(top_sr[['Player', 'Country', 'SR', 'Ball_Faced']])
```

2. Most Consistent Players (Lowest Number of 0s)

```python
# Filter players with minimum innings played
min_innings = 5
consistent_players = df[df['Inning'] >= min_innings].copy()

# Calculate ducks per innings
consistent_players['Ducks_per_Inning'] = consistent_players['0'] / consistent_players['Inning']

# Get most consistent (least ducks)
most_consistent = consistent_players.sort_values(['0', 'Ducks_per_Inning'])[['Player', 'Country', 'Inning', '0', 'Ducks_per_Inning']].head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Ducks_per_Inning', y='Player', data=most_consistent.head(5), palette='Purples_d')
plt.title('Top 5 Most Consistent Players (Lowest Ducks per Inning)')
plt.xlabel('Ducks per Inning')
plt.ylabel('Player')
plt.show()

# Display table
print(f"Most Consistent Players (min {min_innings} innings, lowest ducks):")
display(most_consistent)

# Bonus: Players with most ducks
most_ducks = consistent_players.sort_values(['0', 'Ducks_per_Inning'], ascending=False)[['Player', 'Country', 'Inning', '0', 'Ducks_per_Inning']].head(5)
print("\nPlayers with Most Ducks:")
display(most_ducks)
```

3. Boundary Hitters Analysis (4s and 6s)

```python
# Create boundary metrics
df['Total_Boundaries'] = df['4s'] + df['6s']
df['Boundary_Percentage'] = (df['4s']*4 + df['6s']*6) / df['Runs'] * 100
df['Boundaries_per_Inning'] = df['Total_Boundaries'] / df['Inning']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Top boundary hitters (total)
top_boundaries = df.sort_values('Total_Boundaries', ascending=False).head(5)
sns.barplot(x='Total_Boundaries', y='Player', data=top_boundaries, palette='Oranges_d', ax=axes[0, 0])
axes[0, 0].set_title('Top 5 Players by Total Boundaries (4s + 6s)')

# Top 6 hitters
top_6s = df.sort_values('6s', ascending=False).head(5)
sns.barplot(x='6s', y='Player', data=top_6s, palette='Blues_d', ax=axes[0, 1])
axes[0, 1].set_title('Top 5 Players by 6s')

# Highest boundary percentage (min 100 runs)
min_runs = 100
top_boundary_pct = df[df['Runs'] >= min_runs].sort_values('Boundary_Percentage', ascending=False).head(5)
sns.barplot(x='Boundary_Percentage', y='Player', data=top_boundary_pct, palette='Greens_d', ax=axes[1, 0])
axes[1, 0].set_title(f'Top 5 Players by Boundary % (min {min_runs} runs)')
axes[1, 0].set_xlabel('% of Runs from Boundaries')

# Boundaries per inning (min 5 innings)
min_innings = 5
top_bpi = df[df['Inning'] >= min_innings].sort_values('Boundaries_per_Inning', ascending=False).head(5)
sns.barplot(x='Boundaries_per_Inning', y='Player', data=top_bpi, palette='Reds_d', ax=axes[1, 1])
axes[1, 1].set_title(f'Top 5 Players by Boundaries per Inning (min {min_innings} innings)')

plt.tight_layout()
plt.show()

# Display tables
print("Top 5 Players by Total Boundaries:")
display(top_boundaries[['Player', 'Country', '4s', '6s', 'Total_Boundaries']])

print("\nTop 5 Players by 6s:")
display(top_6s[['Player', 'Country', '6s', 'Total_Boundaries']])

print(f"\nTop 5 Players by Boundary % (min {min_runs} runs):")
display(top_boundary_pct[['Player', 'Country', 'Boundary_Percentage', 'Runs', '4s', '6s']])

print(f"\nTop 5 Players by Boundaries per Inning (min {min_innings} innings):")
display(top_bpi[['Player', 'Country', 'Boundaries_per_Inning', 'Inning', '4s', '6s']])

# Scatter plot of 4s vs 6s
plt.figure(figsize=(10, 6))
sns.scatterplot(x='4s', y='6s', data=df, hue='Country', s=100, alpha=0.7)
plt.title('4s vs 6s Comparison')
plt.xlabel('Number of 4s')
plt.ylabel('Number of 6s')

# Highlight top boundary hitters
highlight_players = top_boundaries['Player'].tolist() + top_6s['Player'].tolist()
for _, row in df[df['Player'].isin(highlight_players)].iterrows():
    plt.text(row['4s']+1, row['6s']+1, row['Player'], fontsize=9)

plt.show()
```

2. Player Clustering (Aggressive, Anchor, Finisher)

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Filter players with sufficient innings
cluster_data = df[df['Inning'] >= 5][['Player', 'Country', 'SR', 'Average']].dropna()

# Scale features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data[['SR', 'Average']])

# Find optimal clusters (Elbow Method)
wcss = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 4))
plt.plot(range(1, 6), wcss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Interpretation
Sharp Decline at 2–3 Clusters: Biggest improvement in grouping quality.

Elbow Point at 3 Clusters: Where the WCSS curve "bends" – diminishing returns beyond this.

Why 3?: Matches cricket’s natural roles: Anchors, Aggressors, Finishers.

Flatline Beyond 3: Adding more clusters (4–5) doesn’t significantly improve grouping.

Strategic Insight
3 Clusters Are Ideal: Aligns with prior analysis of player roles.

Example: Kohli (Anchor), Maxwell (Aggressor), de Kock (Finisher).

Avoid Over-Clustering: 4–5 groups would split roles artificially (e.g., separating "anchors" into defensive/offensive).



```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
cluster_data['Cluster'] = clusters

# Label clusters
cluster_labels = {
    0: 'Anchor (High Avg, Low SR)',
    1: 'Aggressive (High SR, Low Avg)',
    2: 'Finisher (Balanced Avg & SR)'
}
cluster_data['Role'] = cluster_data['Cluster'].map(cluster_labels)

# Display cluster centers (original scale)
centers = scaler.inverse_transform(kmeans.cluster_centers_)
centers_df = pd.DataFrame(centers, columns=['SR', 'Average'])
centers_df['Role'] = ['Anchor', 'Aggressive', 'Finisher']
print("\nCluster Characteristics:")
display(centers_df)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='SR', y='Average', data=cluster_data, hue='Role', palette='Set1', s=100)
plt.title('Player Clusters by Strike Rate and Average')
plt.xlabel('Strike Rate')
plt.ylabel('Batting Average')
plt.show()
```
